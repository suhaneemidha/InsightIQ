```python
import pandas as pd
import os

data_path = r"C:\Users\Snehal\Desktop\InsightIQ\data"

files = os.listdir(data_path)

dfs = {}

for f in files:
    if f.endswith(".csv"):
        file_path = os.path.join(data_path, f)

        name = f.replace(".csv", "")
        dfs[name] = pd.read_csv(file_path)

        print(
            f"Loaded: {name} -> "
            f"{dfs[name].shape[0]} rows, "
            f"{dfs[name].shape[1]} columns"
        )
```

```python
for name, df in dfs.items():
    print(f"{name}: {df.shape}")
```

```python
for name, df in dfs.items():
    nulls = df.isnull().sum().sum()
    print(f"{name}: {nulls} null values")
```

```python
for name, df in dfs.items():
    print(f"\n--- {name} ---")
    print(df.columns.tolist())
```
